## Create the busiest day

```python
# Change "order_dow" column name to "order_day_of_week."

df_large.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
# Find frequency of orders_day_of_week.

df_large['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Use for-loop to determine whether orders are on "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week).

result = []

for value in df_large["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column in df_large to view results in context.

df_large['busiest_day'] = result
```

```python
df_large['busiest_day'].value_counts(dropna = False)
```

### Step 2: Update busiest_day

```python
# Create column busiest_days" from "orders_day_of_week" to identify two busiest days (0, 1), slowest days (4, 3), or regularly busy (all other days).

result_modified = []

for value in df_large["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_modified.append("Busiest days")
  elif value == 4 or value == 3:
   result_modified.append("Slowest days")
  else:
   result_modified.append("Regularly busy")
```

```python
result_modified
```

### Step 3: Check the column value for accuracy

```python
df_large['busiest_day'] = result_modified
```

```python
# Use for-loop to find the frequency of the most hour of the day

result_hour = []

for value in df_large["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_hour.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_hour.append("Fewest orders")
  else:
    result_hour.append("Average orders")
```
